```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD
```

```python
# ORゲート
# ORゲートの入力値　0と1の組み合わせの行列(4, 2)
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
# ORゲートの出力　正解ラベルの行列(4, 1)
T = np.array([[0], [1], [1], [1]])
```

```python
'''
モデルの作成
'''
model = Sequential()     # Sequentialオブジェクトの生成
model.add(Dense(         # 層を追加
    units=1,             # ユニット数は1個のみ
    input_dim=2,         # 入力層のユニット数は2個
    activation='sigmoid' # 活性化関数はシグモイド
))

'''
モデルのコンパイル
'''
model.compile(
    loss='binary_crossentropy', # クロスエントロピー誤差関数はバイナリ専用
    optimizer=SGD(lr=0.1)       # 最適化に勾配降下法を使用
)

# モデルのサマリーを表示
model.summary()
```

```python
'''
モデルの学習
'''
history = model.fit(X,            # 訓練データ
                    T,            # 正解ラベル
                    epochs=200,   # 学習回数
                    batch_size=1, # 確率的勾配効果法で使用するサンプルの数
                    verbose=1     # 学習の進捗状況を出力する
                   )
```

```python
'''
学習結果の確認
'''
# 入力データに学習結果を適用して予測する
classes = model.predict_classes(X, batch_size=1)
print('classified:')
print(T == classes)  # 予測結果と正解ラベルが等しいか
# 最終出力を取得して出力する
prob = model.predict_proba(X, batch_size=1)
print('\noutput probability:')
print(prob)
```
